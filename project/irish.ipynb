```python
import csv
```

```python
f = open('irished.csv', 'rb')
reader = csv.reader(f)
DVRT = []
Type_Of_School=[]

for row in reader:
    a= row[0].split()
    b = a[1:]
    c=  map(int, b)
    DVRT.append(c[0])
    Type_Of_School.append (c[4])

print DVRT
print Type_Of_School

        
f.close()
```

```python
from pylab import *
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.set_title('Irish DVRT and Type_Of_School')
ax.set_xlabel('Irish DVRT(g)')
ax.set_ylabel('Irish Type_Of_School (cc)')
ax.scatter(DVRT, Type_Of_School)
fig.savefig("irish.png")

```

This graph shows the score of The Verbal Reasoning Test compared with the Type of Schools if it was secondary, vocational or primary terminal leaver.
