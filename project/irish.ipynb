```python
import csv
```

```python
f = open('irished.csv', 'rb')
reader = csv.reader(f)
Gender = []
DVRT=[]
A=[]
B=[] 

for row in reader:
    a= row[0].split()
    b = a[0:]
    c=  map(int, b)
    Gender.append(c[0])
    DVRT.append (c[1])
    
if c[0:] = 1 : 
    A.append(c[1:])
else :
    B.append(c[1:])
        

print Gender
print DVRT
print A
print B

        
f.close()
```

```python
from pylab import *
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.set_title('Irish  Gender and DVRT')
ax.set_xlabel('Irish  Gender(1 = male     2= female)')
ax.set_ylabel('Irish DVRT ')
ax.scatter( Gender,DVRT )
fig.savefig("irish.png")

```

This graph shows the score of The Verbal Reasoning Test compared with the Type of Schools if it was secondary, vocational or primary terminal leaver.
